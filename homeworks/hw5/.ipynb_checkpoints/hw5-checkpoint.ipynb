```python
from pymongo import MongoClient
import pymongo
import pandas as pd
import re
client = MongoClient("mongodb://localhost:27017")
client.drop_database("hw5")
```

```python
db = client["hw5"]
movies = [
    {
        'title': 'Demon Slayer',
        'rating': 5,
        'genre': 'Fantasy'
    },
    {
        'title': 'Guardians of the Galaxy',
        'rating': 3,
        'genre': 'Sci-Fi'
    },
    {
        'title': 'Dr.Stone',
        'rating': 5,
        'genre': 'Comedy'
    },
    {
        'title': 'JJK',
        'rating': 2,
        'genre': 'Action'
    },
     {
        'title': 'Your Name',
        'rating': 2,
        'genre': 'Action'
    }

]
myMovies = db['myMovies']
result = myMovies.delete_many({})
result = myMovies.insert_many(movies)
```

```python
#3a
print("3a")
for movie in movies:
    print(movie)
#3b
print("\n3b")
for movie in movies:
    if movie['title'] == 'Demon Slayer':
        print(movie)
#3c
print("\n3c")
sorted_movies = sorted(movies, key=lambda k: k['rating'], reverse=True)
movie_rates = sorted_movies[:3]
for movie in movie_rates:
    print(movie)
```

```python
#5
mov_col = db['movies']
result = mov_col.delete_many({})
rat_col = db['ratings']
result = rat_col.delete_many({})
tag_col = db['tags']
result = tag_col.delete_many({})


df = pd.read_csv('movies.csv')
data = df.to_dict(orient='records')
mov_col.insert_many(data)

df= pd.read_csv('tags.csv')
data = df.to_dict(orient='records')
tag_col.insert_many(data)

df = pd.read_csv('ratings.csv')
data = df.to_dict(orient='records')
rat_col.insert_many(data)
```

```python



for doc in mov_col.find():
    title = doc["title"]
    match = re.search(r'\((\d{4})\)', title)
    if match:
        year = match.group(1)
        mov_col.update_one({"_id": doc["_id"]}, {"$set": {"year": int(year)}})
```

```python
#6a
print("\nDevelop code to find number of movies released per year")

pl = [{"$group": {"_id": { "year": "$year" },"count": {"$sum": 1}}}]

result = mov_col.aggregate(pl)
for doc in result:
    print(doc)
```

```python

print("\nDevelop code to find number of movies per genre")
pl = [{"$addFields":{"genres": {"$split": ["$genres", "|"]}}},{"$unwind": "$genres"},{"$group": {"_id": "$genres","count": {"$sum": 1}}}]

result = mov_col.aggregate(pl)

for doc in result:
    print(doc)
```

```python
print("\nDevelop code to find number of movies per rating")
pl = [{"$group": {"_id": "$rating", "count": {"$sum": 1}}}]

result = list(rat_col.aggregate(pl))

for doc in result:
    print(f"Rating: {doc['_id']}, Number of movies: {doc['count']}")
    
```

```python

print("\nDevelop code to find number of movies tagged")
pl = [{"$unwind": "$tag"},{"$group": {"_id": "$tag", "count": {"$sum": 1}}}]

result = tag_col.aggregate(pl)

for doc in result:
    print(doc)

```
